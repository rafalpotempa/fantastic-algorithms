# Bogosort (random permutations)

Source: Gruber H., Holzer M., Ruepp O., *Sorting the Slow Way: An Analysis of Perversely Awful Randomized Sorting Algorithms*, 2007

Complexity: $O(n \cdot n!)$

```python
from random import shuffle
from visualization.bogosort import visualize
```

```python
@visualize
def bogosort(array):
    result = array
    while True:
        shuffle(array)
        if array == sorted(array):
            break
        yield array
    yield result

bogosort([2, 4, 5, 2, 1, 7])
```
