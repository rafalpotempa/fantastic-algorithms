# Fibonacci numbers by matrix multiplication

Source: S. Dasgupta, C. Papadimitriou, U. Vazirani, *Algorithms*, 2008

Complexity: $O(\log{}n)$

```python
import numpy as np
```

```python
X = np.array([[0, 1],
              [1, 1]])
F01 = np.array([[0],
                [1]])

def fib_matrix(n):
	X_to_n = X
	for _ in range(n):
		X_to_n = X_to_n @ X
	return (X_to_n @ F01)[0,0]
```

```python
', '.join([str(fib_matrix(i)) for i in range(20)])
```
